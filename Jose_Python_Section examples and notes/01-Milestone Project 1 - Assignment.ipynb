```python
from IPython.display import clear_output
clear_output()
from random import randrange
i_num = 0
i_list = []
i_example = []
char_Value = [" ", " "]
i_xcount = 0
i_ocount = 0
s_player_string = f" Player 1"
def char_select():
    global i_player
    global char_Value
    global s_player_string
    i_player = randrange(1,3)
    s_player_string = f"Player {i_player}"
    b_continue = True
    usr_choice = ""
    while b_continue:
        clear_output()
        usr_choice = input (f' This is a two player game are the players ready? Y/N \n')
        if (usr_choice.lower() == "y" or usr_choice.lower() == 'yes'):
            b_continue = False
    b_continue = True
    usr_choice = ""
    while b_continue:
        usr_choice = input (f' {s_player_string} do you wish to be "X" or "O"? \n')
        if ( usr_choice.lower() == "x" or usr_choice.lower() == "o"):
            b_continue = False
            if (usr_choice.lower() == 'x' and i_player == 1 ) or (usr_choice == "o" and i_player == 2):
                char_Value = ["X","O"]
            else:
                char_Value = ["O", "X"]
        else:
            b_continue = True
def game_board():
    '''
    Display the game board and clear it after every choice
    '''
    clear_output()
    global i_player
    global i_list
    global i_example
    global char_Value
    global s_player_string
    i_line = [" "]*6
    j_line = [" "]*6
    j = 0
    for i in range (6,-1,-3):
        i_line[j] = f" {i_example[i]} | {i_example[i+1]} | {i_example[i+2]}"
        i_line[j+1] = "_"*12
        j +=2
    print (f' Game board layoyt: ')
    for i in i_line:
        print(i)
    j = 0
    for i in range (6,-1,-3):
        j_line[j] = f" {i_list[i]} | {i_list[i+1]} | {i_list[i+2]}"
        j_line[j+1] = "_"*12
        j +=2
    print (f" {s_player_string} it is your turn you are {char_Value[i_player -1]}'s")
    for i in j_line:
        print(i)
def user_choice():
    '''
    User inputs a number (0-10) and we return this in integer form. This will verify that it is a number and with in the correct range
    '''
    global s_text
    B_Error = True
    i_range = range(1,10)
    while (B_Error):
        try:
            result = int(input(f' {s_text} \n'))
            B_Error = False
            if (not result in i_range):
                B_Error= True
                clear_output()
                game_board()
                print('Error outside the ranage')
        except ValueError:
            clear_output()
            game_board()
            print (f'Error {s_text}')
            B_Error = True
    return result
def board_check(choice):
    '''
    This will verify the number selection has not been used yet
    '''
    global i_list 
    global i_example
    global s_text
    global i_player
    global i_turn
    global char_Value
    global s_player_string
    s_text = f"Please enter a numerical value between 0 and 10 : See Game board layout"
    try:        
        if (not choice in i_list):
            # removes the number from the example game and puts the players X/O on the board
            i_example[i_example.index(choice)] = " "
            i_list[choice-1] = char_Value[i_player-1]
            i_turn += 1
            if (i_player == 1):
                i_player +=1
                s_player_string = f" Player {i_player}"
            else:
                i_player = 1
                s_player_string = f" Player {i_player}"
            return False
    except ValueError:
        clear_output()
        s_text = f"{s_player_string} Please enter a numerical value between 0 and 10 : your last choice has already been used See Game board layout" 
        game_board()
        return True
def game_counter (i_list, loopcount : 3):
    global i_player
    global j
    global i_xcount
    global i_ocount
    global s_player_string
    j += 1
    if (i_list == "X"):
        i_xcount +=1
    elif (i_list == "O"):
        i_ocount +=1
# This would denote a winner remove shifting player back to the winning player else leve alone and continue game
    if (i_xcount == 3 or i_ocount == 3):
        if (i_player == 1):
            i_player = 2
            s_player_string = f" Player {i_player}"
        else:
            i_player = 1
            s_player_string = f" Player {i_player}"
        return True
    if (j >= loopcount):
        i_xcount = 0
        i_ocount = 0
        j = 0
def game_check (i_list):
    global j
    j = 0
    for i in range (0,3,1):
        if game_counter(i_list[i],3):
            return False
    for i in range (3,6,1):
        if game_counter(i_list[i],3):
            return False
    for i in range (6,9,1):
        if game_counter(i_list[i],3):
            return False
    for i in range (0,9,3):
        if game_counter(i_list[i],3):
            return False
    for i in range (1,9,3):
        if game_counter(i_list[i],3):
            return False
    for i in range (2,9,3):
        if game_counter(i_list[i],3):
            return False
    for i in range (0,9,4):
        if game_counter(i_list[i],3):
            return False
    for i in range (2,7,2):
        if game_counter(i_list[i],3):
            return False
    return True
```

```python
b_PlayGame = True
while (b_PlayGame):
    # Set base falgs at at start of game and clear board
    b_game_continue = True
    b_continue = True
    b_Error_BC = True
    s_exit = f"{s_player_string} is the winner!'\n' Do you wish to play again Y/N?"
    # Set the initail board that will be removed as characters use the numbers
    i_turn = 0
    s_text = f"Please enter a numerical value between 0 and 10 : See Game board layout"
    i_example = [1,2,3,4,5,6,7,8,9]
    i_list = [" "]*9
    char_select()
    while (b_game_continue):
        game_board()
        while (b_Error_BC):
            choice = user_choice()
            b_Error_BC = board_check(choice)
        b_Error_BC = True
        if (i_turn > 3):
            b_game_continue = game_check(i_list)
        elif(i_turn >=8):
            b_game_continue = false
            s_exit = f"The game was a draw do you wish to play again Y/N?"
    while (b_continue):
        game_board()
        usr_choice = input (s_exit)
        if (usr_choice.lower() == "y" or usr_choice.lower() == 'yes'):
            clear_output()
            b_continue = False
        elif(usr_choice.lower() == "n" or usr_choice.lower() == 'no'):
            b_PlayGame = False
            b_continue = False
            b_game_continue = False
            print ("Exiting Game")
        else:
            clear_output()
            print ("Invalid entry try again")
# End of code loop back to top or exit the game
# Refresh the board to show the winning board
```
